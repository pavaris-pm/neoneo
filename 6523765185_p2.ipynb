```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
data = pd.read_csv('/content/drive/MyDrive/datasets/supermarket.csv')
data.head()
```

```python
# remove the column that is non-binary
pick_cols = [col for col in data.columns if col != "receipt_id"]
data = data[pick_cols]
data.head()
```

```python
# Apply Apriori algorithm to find frequent itemsets
frequent_itemsets = apriori(data, min_support=0.2, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```

```python
# Apply Apriori algorithm to find frequent itemsets with lower minimum support
frequent_itemsets = apriori(data, min_support=0.1, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules
```
